```python
cn.sort_values(by='Date')
cn[cn['City']=='Beijing']['median'].isnull().sum()
sorted(cn[cn['Specie']=='aqi']['Date'].unique())
```

```python
cn_gust = cn[cn['Specie']=='wind-gust'][['City', 'Date', 'median']]
cn_gust2 = cn[cn['Specie']=='wind gust'][['City', 'Date', 'median']]
cn_ws = cn[cn['Specie']=='wind-speed'][['City', 'Date', 'median']]
cn_ws2 = cn[cn['Specie']=='wind speed'][['City', 'Date', 'median']]
cn_hum = cn[cn['Specie']=='humidity'][['City', 'Date', 'median']]
cn_perc = cn[cn['Specie']=='precipitation'][['City', 'Date', 'median']]
cn_dew = cn[cn['Specie']=='dew'][['City', 'Date', 'median']]
cn_pm25 = cn[cn['Specie']=='pm25'][['City', 'Date', 'median']]
```

```python
cn_gust.rename(columns={'median':'gust'}, inplace=True)
cn_gust2.rename(columns={'median':'gust'}, inplace=True)
cn_hum.rename(columns={'median':'hum'}, inplace=True)
cn_perc.rename(columns={'median':'perc'}, inplace=True)
cn_dew.rename(columns={'median':'dew'}, inplace=True)
cn_ws.rename(columns={'median':'ws'}, inplace=True)
cn_ws2.rename(columns={'median':'ws'}, inplace=True)
cn_pm25.rename(columns={'median':'pm25'}, inplace=True)
cn_gust.append(cn_gust2)
cn_ws.append(cn_ws2)
cn_pm25
```

```python
cn_concat = pd.merge(cn_ws, cn_hum, how='left', left_on = ['City', 'Date'], right_on = ['City', 'Date'])
# cn_concat = pd.merge(cn_concat, cn_perc, how='left')
cn_concat = pd.merge(cn_concat, cn_dew, how='left', left_on = ['City', 'Date'], right_on = ['City', 'Date'])
cn_concat = pd.merge(cn_concat, cn_pm25, how='left', left_on = ['City', 'Date'], right_on = ['City', 'Date'])
cn_concat
```
